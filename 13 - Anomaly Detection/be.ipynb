```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
df = pd.read_csv('data/dataset.csv')
```

```python
df.head()
```

```python
def get_day (day_cycle_window) : 
    splitted = day_cycle_window.split('_')
    return int(splitted[0])

def get_cycle (day_cycle_window) : 
    splitted = day_cycle_window.split('_')
    return int(splitted[1])

def get_window (day_cycle_window) : 
    splitted = day_cycle_window.split('_')
    return int(splitted[2])

days = df['day_cycle_window'].apply(get_day)
df = df.assign(Day=days)
cycle = df['day_cycle_window'].apply(get_cycle)
df = df.assign(Cycle=cycle)
window = df['day_cycle_window'].apply(get_window)
df = df.assign(Window=window)


df_features = df.groupby(['day_cycle_window']).mean()
```
